# Ridge Regression Lasso Regression Elastic net

```python
from sklearn.datasets import load_boston
boston=load_boston()
boston
```

# STEPS


```python
-->Data calling
-->x,y
-->From sklearn ,odel calling
-->train,test
-->Data fitting to the model
-->Predict(testing)
-->Score(train,testing)
-->plotting
```

```python
boston.data
```

```python
boston.target
```

```python
boston.feature_names
```

```python
#creating dataset
import pandas as pd
boston_df=pd.DataFrame(boston.data,columns=boston.feature_names)
```

```python
boston_df.head()
```

```python
boston_df["Land Price"]=boston.target
```

```python
boston_df.head()
```

```python
x=boston_df.drop("Land Price",axis=1)
x
```

```python
x=boston_df.drop("Land Price",axis=1)
x
```

```python
x.values
```

```python
y=boston_df["Land Price"]
y.values
```

```python
#Ridge Regression=sumof error+sum of squares of magnitude
#coefficient(alpha)
from sklearn.linear_model import Ridge
rr=Ridge(alpha=1,solver='svd')
rr
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=2,random_state=3)
```

```python
rr.fit(x_train,y_train)
```

```python
y_predict=rr.predict(x_train)
y_predict
```

```python
#score
rr.score(x_train,y_train)
```

```python
rr.score(x_test,y_test)
```

```python
rr.coef_
```

```python
from sklearn.linear_model import LinearRegression
lin=LinearRegression()
```

```python
lin.fit(x_train,y_train)
```

```python
train_score=lin.score(x_train,y_train)
```

```python
print("Lin Train score",train_score)
```

```python
test_score=lin.score(x_test,y_test)
print("Lin Testing score",test_score)
```

```python
lin.coef_
```

```python
import matplotlib.pyplot as plt
plt.plot(rr.coef_,marker="*",label="Ridge Coefficient",color="y")
plt.plot(lin.coef_,marker=".",label="Linear Coefficient",color="r")
plt.xlabel("coefficient index")
plt.ylabel("coefficient magnitude")
plt.title("Ridge vs Linear")
plt.legend()
plt.show()
```
