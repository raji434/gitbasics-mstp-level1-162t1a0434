```python
#creating subplot using numpy with sine
import numpy as np
t=np.arange(0.0,0.5,0.1)
s=np.sin(2*np.pi*t)
plt.subplot(2,2,3)
plt.plot(t,s,'rx',label="1st graph with sine wave")
plt.xlabel("t")
plt.ylabel("s")
plt.title("graph t vs s")
plt.legend()
t[0]=np.nan
t[1]=np.nan
plt.subplot(2,1,2)
plt.plot(t,s,label="2nd graph with sine wave")
plt.xlabel("t")
plt.ylabel("s")
plt.legend()
plt.show()
```

```python
x=[1,3,2,4]
y=[3,5,4,2]
z=[2,5,3,4]
fig=plt.figure()
ax=fig.add_subplot(212,projection='3d')
ax.plot(x,y,z)
plt.show()
plt.xlabel("x")
plt.ylabel("y")
plt.title("3d projection")
plt.show()
```

```python
x=[1,3,2,4]
y=[3,5,4,2]
z=[2,5,3,4]
dx=np.ones(4)
dy=np.zeros(4)
dz=np.ones(4)
fig=plt.figure()
ax=fig.add_subplot(212,projection='3d')
ax.bar3d(x,y,z,dx,dy,dz,color='m')
plt.xlabel("x")
plt.ylabel("y")
plt.title("3d projection")
plt.show()
```
