```python
import numpy as np
import matplotlib.pyplot as plt
x=np.random.rand(20)
y=np.random.rand(20)
plt.plot(x,y,color="k",alpha=0.1,linewidth=2,markersize=20)
```

```python
help(plt.plot)
```

```python
help(sns.set)
```

```python
sns.set()
plt.plot(x,y)
```

```python
sns.set(color_codes=True)
plt.plot(x,y)
```

```python
sns.set(palette="dark")
plt.plot(x,y)
```

```python
sns.set(style='darkgrid')
t=sns.load_dataset("tips")
t.head()
```

```python
sns.set(style='darkgrid')
t=sns.load_dataset("tips")
t.tail()
```

```python
sns.relplot(x="total_bill",y="day",data=t)
```

```python
sns.relplot(x="total_bill",y="day",data=t,hue="tip")
```

```python
sns.set(style='darkgrid')
f=sns.load_dataset("flights")
f.head()
```

```python
sns.set(style='darkgrid')
f=sns.load_dataset("flights")
f.tail()
```

```python
sns.relplot(x='year',y='passengers',data=f,hue="month",col="month")
```
